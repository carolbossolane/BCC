1) Para este item você deverá usar o arquivo com a porcentagem da população brasileira que é analfabeta, dividida por unidade da federação, de 1981 a 2014 (analfabetismo.csv, encontrado na pasta aula5/dados) (dados disponibilizados por Ipeadata).

Faça um gráfico da taxa de analfabetismo em São Paulo em função do tempo (anos).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

populacao = pd.read_csv('https://raw.githubusercontent.com/BMClab/BasesComputacionais/master/dados/analfabetismo.csv')

plt.figure()
plt.plot(populacao['São Paulo'], populacao['Ano'])
plt.grid()
plt.xlabel('Taxa de analfabetismo')
plt.ylabel('Ano')
plt.show()

```
