```python
#1)
import numpy as np 
import matplotlib.pyplot as plt

t = np.arange(0 , 10.05 , 0.05)
v = np.exp(-0.5*t) * np.cos(2*np.pi*0.8*t)

plt.figure()
plt.plot(t, v)
plt.xlabel('tempo(s)')
plt.ylabel('tensão(volts)')
plt.title('Gráfico da tensão em função do tempo')
plt.show()


```

```python
#2)
import numpy as np 
import matplotlib.pyplot as plt

C = np.linspace(-20,100,100)
F = 9/5*C + 32

plt.figure()
plt.plot(C, F)
plt.xlabel('celsius')
plt.ylabel('fahrenheit')
plt.show()
```

```python
#3)
import numpy as np 
import matplotlib.pyplot as plt

a = np.arange(1994,2022)
P = 800000*np.exp((a-1994)/40)

plt.figure()
plt.plot(a, P)
plt.xlabel('anos')
plt.ylabel('população')
plt.show()
```
