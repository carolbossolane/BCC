1) Para este item você deverá usar o arquivo com dados de força e torque medidos no chão enquanto uma pessoa tenta ficar em pé durante um minuto.

Faça o gráfico do torque na direção y (My [Nm]) em função do tempo (Time [s]).

Mostre na tela a média e o desvio-padrão do torque na direção y.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

tabela= pd.read_csv("https://raw.githubusercontent.com/joaozz21/BasesComputacionais/master/aula7/dados/balance.csv")
my= tabela['My[Nm]']
time= tabela['Time[s]']
media= my.mean()
desvio= my.std()
print(type(media))
plt.figure()
plt.plot(time, my, label='torque', color = 'black')
plt.axhline(media, marker = '', linestyle = '-', color = 'red', label='média')
plt.axhline(media + desvio, marker = '', linestyle = '--', color = 'blue', label='desvio-padrão')
plt.axhline(media - desvio, marker = '', linestyle = '--', color = 'blue')
plt.grid()
plt.legend()
plt.xlabel('Tempo')
plt.ylabel('Torque na direção y')
plt.show()
```

2) Para este item você deverá usar o arquivo com a taxa de inflação mensal (IGP-DI) no Brasil de fevereiro de 1944 a junho de 2019 (dados disponibilizados por Ipeadata).

Calcule a mediana da inflação mensal nos meses de março.

Calcule a média da inflação mensal nos meses de março.

Calcule o desvio-padrão da inflação mensal nos meses de março.

Faça o histograma da inflação mensal nos meses de março.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

tabela= pd.read_csv("https://raw.githubusercontent.com/joaozz21/BasesComputacionais/master/aula7/dados/inflacaoMensal.csv")
março= tabela.query('Mês == 3')

#a)
março_sort=março['Inflação'].sort_values()
mediana=março_sort.median()
print('A mediana da inflação mensal nos meses de Março é', mediana)

#b)
media=março_sort.mean()
print('A média da inflação mensal nos meses de Março é', round(media,2))

#c)
desvio=março_sort.std()
print('O desvio-padrão da inflação mensal nos meses de Março é', round(desvio,2))

#d)
plt.figure()
plt.grid()
março['Inflação'].hist()
plt.xlabel('inflação mensal nos meses de março')
plt.show()
```

4) Use a sequência de instruções abaixo para gerar N números aleatórios entre 0 e 1.

rng = np.random.default_rng()
x = rng.uniform(size=N)

Faça o histograma e calcule (e mostre na tela) a média dos valores gerados.

Faça isso para a) N = 100, N = 1000, c) N = 10000, d) N = 100000

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#a)
N=100
rng = np.random.default_rng()
x = rng.uniform(size=N)
tabela= pd.DataFrame(x)
media= tabela[0].mean()

plt.figure()
plt.grid()
tabela[0].hist()
plt.axhline(media, marker = '', linestyle = '-', color = 'red', label='média')
plt.legend()
plt.show()

#b)
N=1000
rng = np.random.default_rng()
x = rng.uniform(size=N)
tabela= pd.DataFrame(x)
media= tabela[0].mean()

plt.figure()
plt.grid()
tabela[0].hist()
plt.axhline(media, marker = '', linestyle = '-', color = 'red', label='média')
plt.legend()
plt.show()

#c)
N=10000
rng = np.random.default_rng()
x = rng.uniform(size=N)
tabela= pd.DataFrame(x)
media= tabela[0].mean()

plt.figure()
plt.grid()
tabela[0].hist()
plt.axhline(media, marker = '', linestyle = '-', color = 'red', label='média')
plt.legend()
plt.show()

#d)
N=100000
rng = np.random.default_rng()
x = rng.uniform(size=N)
tabela= pd.DataFrame(x)
media= tabela[0].mean()

plt.figure()
plt.grid()
tabela[0].hist()
plt.axhline(media, marker = '', linestyle = '-', color = 'red', label='média')
plt.legend()
plt.show()
```
